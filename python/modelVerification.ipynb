```python
import pandas as pd
import numpy as np
import statistics
import matplotlib.pyplot as plt

data = pd.read_csv("ISS-TEST_data.txt", sep="\t")

RA = data["        R.A., deg"].to_numpy()
Dec = data["          Dec, deg"].to_numpy()
Az = data["        Az, deg"].to_numpy()
Elev = data["        Elev, deg"].to_numpy()
MJD = data["          MJD"].to_numpy()



data_Kostaya = pd.read_csv("resultsISSKostayaTab.txt", sep="\t")

RA_Kostaya = data_Kostaya["RA"].to_numpy()
Dec_Kostaya = data_Kostaya["DE"].to_numpy()
Az_Kostaya = data_Kostaya["AZI"].to_numpy()
Elev_Kostaya = data_Kostaya["ELE"].to_numpy()
MJD_Kostaya = data_Kostaya["Time(MJD)"].to_numpy()


RA_Kostaya = np.delete(RA_Kostaya, -1)
Dec_Kostaya = np.delete(Dec_Kostaya, -1)
Az_Kostaya = np.delete(Az_Kostaya, -1)
Elev_Kostaya = np.delete(Elev_Kostaya, -1)
MJD_Kostaya = np.delete(MJD_Kostaya, -1)


#calculate differences between data
dRA = np.absolute(RA - RA_Kostaya)
dDec = np.absolute(Dec - Dec_Kostaya)
dAz = np.absolute(Az - Az_Kostaya)
dElev = np.absolute(Elev - Elev_Kostaya)
dMJD = np.absolute(MJD - MJD_Kostaya)

dRA_mean = statistics.mean(dRA)
dDEC_mean = statistics.mean(dDec)
dAz_mean = statistics.mean(dAz)
dElev_mean = statistics.mean(dElev)
dMJD_mean = statistics.mean(dMJD)

dRA_variance = statistics.variance(dRA)
dDEC_variance = statistics.variance(dDec)
dAz_variance = statistics.variance(dAz)
dElev_variance = statistics.variance(dElev)
dMJD_variance = statistics.variance(dMJD)

print("dRA mean", dRA_mean, '\n', "dDEC mean", dDEC_mean, '\n', "dAz mean", dAz_mean, '\n', "dElev mean", dElev_mean, '\n', 'dMJD mean', dMJD_mean)
print("dRA var", dRA_variance, '\n', "dDEC var", dDEC_variance, '\n', "dAz var", dAz_variance, '\n', "dElev var", dElev_variance, '\n', 'dMJD var', dMJD_variance)

ax1 = plt.subplot(2, 2, 1)
ax1.plot(range(len(dRA)),dRA)
ax1.set(xlabel='', ylabel='dRA mean, deg')
ax1.grid()

#plot 2:
ax1 = plt.subplot(2, 2, 2)
ax1.plot(range(len(dDec)),dDec)
ax1.set(xlabel='', ylabel='dDec mean, deg')
ax1.grid()

#plot 3:
ax1 = plt.subplot(2, 2, 3)
ax1.plot(range(len(dAz)),dAz)
ax1.set(xlabel='', ylabel='dAz mean, deg')
ax1.grid()

#plot 4:
ax1 = plt.subplot(2, 2, 4)
ax1.plot(range(len(dElev)),dElev)
ax1.set(xlabel='', ylabel='dElev mean, deg')
ax1.grid()

plt.subplots_adjust(wspace=0.4, hspace=0.4)  # Increase spacing

plt.show()


# ax.plot(t_span_obs_day , np.transpose(sunElevationArray_deg))
# plt.axvline(x = observation_time_start, color = 'r', label = 'Observation time start')
# ax.set(xlabel='time, UTC', ylabel='Sun elevation, deg',
#        title='sun elevation for the observation time span')
# ax.grid()

# # fig.savefig("sun elevation.png")
# plt.show()
```
